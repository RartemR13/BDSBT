```python
import json
import redis

redis_db = redis.Redis(host='localhost', port=6380, db=1)
```

```python
with open('users.json') as file:
    data = json.load(file)

print(len(data))
print(data[1])
```

```python
%%time 

pipe = redis_db.pipeline()
for user in data:
    for field in ('age', 'name', 'gender', 'favoriteFruit'):
        pipe.set(f'users:{user["index"]}:{field}', user[field])

pipe.execute()
```

```python
%%time

redis_db.get('users:126:gender')
```

```python
%%time 

pipe = redis_db.pipeline()
for user in data:
    for field in ('age', 'name', 'gender', 'favoriteFruit'):
        pipe.hset(f'users:{user["index"]}', field, user[field])

pipe.execute()
```

```python
%%time

redis_db.hget('users:41279', 'name')
```

```python
%%time 

pipe = redis_db.pipeline()
for user in data:
        pipe.zadd('users_fruits', {user['index']: user['favoriteFruit']})

pipe.execute()
```
